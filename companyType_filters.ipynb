```
org_desc = pd.read_csv('full_data.csv', encoding='utf8')
not_null = org_desc[org_desc.description.notnull()]
```

```
not_null = not_null[not_null.companyType.isnull()]
not_null
```

```
#update companyType if description contains certain words
#higher priority words are at the bottom so they override the top ones
searchfor1 = ['government research', 'government group', 'government']
not_null.loc[not_null['description'].str.contains('|'.join(searchfor1), case = False), 'companyType'] = 'Government Research Group'

searchfor2 = ['non-profit', 'non profit', 'nonprofit','NGO']
not_null.loc[not_null['description'].str.contains('|'.join(searchfor2), case = False), 'companyType'] = 'Non-Profit'

not_null.loc[not_null['description'].str.contains('medical device', case = False), 'companyType'] = 'Medical Device'

searchfor3 = ['pharmaceutical', 'pharma']
not_null.loc[not_null['description'].str.contains('|'.join(searchfor3), case = False), 'companyType'] = 'Pharmaceutical'


searchfor4 = ['biopharmaceutical', 'biopharma']
not_null.loc[not_null['description'].str.contains('|'.join(searchfor4), case = False), 'companyType'] = 'Biopharmaceutical'

not_null.loc[not_null['description'].str.contains('therapeutics', case = False), 'companyType'] = 'Therapeutics'


not_null.loc[not_null['description'].str.contains('diagnostics', case = False), 'companyType'] = 'Diagnostics'

searchfor5 = ['academic research group']  #university for name check
not_null.loc[not_null['description'].str.contains('|'.join(searchfor5), case = False), 'companyType'] = 'Academic Research Group'


searchfor6 = ['contract research', 'CRO','contract research organization']
not_null.loc[not_null['description'].str.contains('|'.join(searchfor6), case = False), 'companyType'] = 'Contract Research (CRO)'


searchfor7 = ['contract manufacturing', 'CMO']
not_null.loc[not_null['description'].str.contains('|'.join(searchfor7), case = False), 'companyType'] = 'Contract Manufacturing (CMO)'


searchfor8 = ['contract development', 'Contract Development and Manufacturing','CDMO','CDO']
not_null.loc[not_null['description'].str.contains('|'.join(searchfor8), case = False), 'companyType'] = 'Contract Development and Manufacturing (CDMO)'
```

```
#same idea here but with names instead of descriptions - names >> descriptions
not_null.loc[not_null['name'].str.contains('therapeutics', case = False), 'companyType'] = 'Therapeutics'
not_null.loc[not_null['name'].str.contains('diagnostics', case = False), 'companyType'] = 'Diagnostics'
not_null.loc[not_null['name'].str.contains('pharma', case = False), 'companyType'] = 'Biopharmaceutical'
not_null.loc[not_null['name'].str.contains('university', case = False), 'companyType'] = 'Academic Research Group'
searchforVC = ['invest', 'investment', 'capital', 'venture']
not_null.loc[not_null['name'].str.contains('|'.join(searchforVC), case = False), 'companyType'] = 'Venture Capital'
```
